```python
import pandas as pd
from io import BytesIO
import requests

url = "https://ftp.ibge.gov.br/Projecao_da_Populacao/Projecao_da_Populacao_2024/projecoes_2024_tab1_idade_simples.xlsx"

# Baixar e carregar planilha com cabeçalho correto
r = requests.get(url)
xls = pd.ExcelFile(BytesIO(r.content))
df = pd.read_excel(xls, sheet_name="1) POP_IDADE SIMPLES", header=5)

# Filtrar apenas "Ambos os sexos"
df = df[df["SEXO"] == "Ambos"]

# Filtrar idades 0 a 14
df = df[df["IDADE"].between(0, 14)]

# Lista das 27 siglas oficiais dos estados
ufs = [
    "AC","AL","AP","AM","BA","CE","DF","ES","GO","MA","MT","MS","MG",
    "PA","PB","PR","PE","PI","RJ","RN","RS","RO","RR","SC","SP","SE","TO"
]

# Selecionar apenas linhas dos estados
df = df[df["SIGLA"].isin(ufs)]

# Selecionar apenas ano 2019
df_2019 = df[["SIGLA", "LOCAL", "IDADE", 2019]]

# Somar população 0 a 14 por estado
pop_0_14_2019 = df_2019.groupby(["SIGLA", "LOCAL"])[2019].sum().reset_index()

# Ordenar do maior para o menor
pop_0_14_2019 = pop_0_14_2019.sort_values(by=2019, ascending=False)

# Exibir resultado final
print("População de 0 a 14 anos por estado (2019):")
display(pop_0_14_2019)

# Soma total e média apenas dos 27 estados
soma_total_2019 = pop_0_14_2019[2019].sum()
media_estados_2019 = pop_0_14_2019[2019].mean()

print("\nSoma total (27 estados, 0-14 anos, 2019):", f"{soma_total_2019:,}".replace(",", "."))
print("Média por estado:", f"{media_estados_2019:,.0f}".replace(",", "."))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Dados
criancas = {
    "Acre": 221160, "Alagoas": 711691, "Amapá": 198176, "Amazonas": 1075426,
    "Bahia": 2863071, "Ceará": 1802435, "Distrito Federal": 534233,
    "Espírito Santo": 741496, "Goiás": 1431289, "Maranhão": 1645580,
    "Minas Gerais": 3707919, "Mato Grosso do Sul": 606792, "Mato Grosso": 829686,
    "Pará": 1990942, "Paraíba": 826921, "Pernambuco": 1894679, "Piauí": 681934,
    "Paraná": 2195297, "Rio de Janeiro": 2856146, "Rio Grande do Norte": 654658,
    "Rio Grande do Sul": 1906510, "Santa Catarina": 1423210, "Sergipe": 469501,
    "São Paulo": 7975163, "Tocantins": 350986, "Rondônia": 348152, "Roraima": 186208
}

mortes = {
    "Paraná": 19440, "Maranhão": 19060, "Ceará": 17620, "Amazonas": 15080,
    "Rio Grande do Sul": 14780, "Goiás": 14340, "Pernambuco": 11374,
    "Santa Catarina": 11280, "Mato Grosso": 10800, "São Paulo": 9573,
    "Alagoas": 9280, "Paraíba": 9020, "Piauí": 7820, "Espírito Santo": 7800,
    "Sergipe": 7560, "Mato Grosso do Sul": 6700, "Rio Grande do Norte": 5600,
    "Distrito Federal": 5100, "Minas Gerais": 4371, "Rio de Janeiro": 4368,
    "Roraima": 4340, "Bahia": 4340, "Tocantins": 3800, "Pará": 3727,
    "Rondônia": 3720, "Amapá": 3400, "Acre": 3320
}

renda = {
    'Acre': 1113, 'Alagoas': 906, 'Amapá': 1066, 'Amazonas': 1020,
    'Bahia': 1139, 'Ceará': 942, 'Distrito Federal': 3174, 'Espírito Santo': 1803,
    'Goiás': 1590, 'Maranhão': 782, 'Mato Grosso': 1689, 'Mato Grosso do Sul': 1894,
    'Minas Gerais': 1635, 'Pará': 973, 'Paraíba': 1129, 'Paraná': 2014,
    'Pernambuco': 1180, 'Piauí': 1032, 'Rio de Janeiro': 3210,
    'Rio Grande do Norte': 1284, 'Rio Grande do Sul': 2210, 'Rondônia': 1717,
    'Roraima': 1291, 'Santa Catarina': 2142, 'São Paulo': 3202,
    'Sergipe': 1187, 'Tocantins': 1055.60
}

# Criar DataFrames separados
df_criancas = pd.DataFrame.from_dict(criancas, orient='index', columns=['Crianças'])
df_mortes = pd.DataFrame.from_dict(mortes, orient='index', columns=['Mortes'])
df_renda = pd.DataFrame.from_dict(renda, orient='index', columns=['Renda Média'])

# Juntar os DataFrames pelo índice (Estado)
df = df_criancas.join(df_mortes).join(df_renda)

# Calcular taxa de mortalidade infantil por 1000 crianças
df['Taxa Mortes por 1000 Crianças'] = (df['Mortes'] / df['Crianças']) * 1000

# Resetar índice para coluna Estado
df = df.reset_index().rename(columns={'index':'Estado'})

# Regressão linear para linha de tendência
slope, intercept, r_value, p_value, std_err = linregress(df['Renda Média'], df['Taxa Mortes por 1000 Crianças'])

# Cores para pontos: vermelho se mortalidade alta (> média), verde se baixa (<= média)
media_mortalidade = df['Taxa Mortes por 1000 Crianças'].mean()
colors = ['red' if x > media_mortalidade else 'green' for x in df['Taxa Mortes por 1000 Crianças']]

# Plot
plt.figure(figsize=(12,7))
plt.scatter(df['Renda Média'], df['Taxa Mortes por 1000 Crianças'], c=colors, s=100, alpha=0.7, edgecolors='k')

# Linha de tendência
x_vals = np.array([df['Renda Média'].min(), df['Renda Média'].max()])
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color='blue', lw=2, label='Linha de tendência')

# Texto com coeficiente de correlação
plt.text(df['Renda Média'].min(), max(df['Taxa Mortes por 1000 Crianças']) * 0.9,
         f"Correlação: {r_value:.2f}", fontsize=14, bbox=dict(facecolor='white', alpha=0.6))

# Texto indicando se renda influencia mortes (baseado na correlação)
influencia = "SIM" if r_value < -0.5 else "NÃO"
plt.text(df['Renda Média'].min(), max(df['Taxa Mortes por 1000 Crianças']) * 0.8,
         f"A renda influencia as mortes? {influencia}", fontsize=14, bbox=dict(facecolor='white', alpha=0.6))

plt.xlabel('Renda Média (R$) - 2019')
plt.ylabel('Taxa de Mortes por 1000 Crianças')
plt.title('Correlação entre Renda Média e Mortalidade Infantil (0-14 anos)')
plt.grid(True)
plt.legend()
plt.show()
```
